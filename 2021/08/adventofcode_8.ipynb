```python
def RCIL(input_str,char_list):
    result = input_str
    for k in char_list:
        result = result.replace(k,'')
    return result
```

```python
# dddd
#e    a
#e    a
# ffff
#g    b
#g    b
# cccc

def asciiNumber(number,poschar):
    result = ''
    if number == 8:
        result += ' '+4*poschar[0]+' \n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += ' '+4*poschar[3]+' \n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += ' '+4*poschar[5]+' \n'

    if number == 7:
        result += ' '+4*poschar[0]+' \n'
        result += '     ' + poschar[2]+'\n'
        result += '     ' + poschar[2]+'\n'
        result += '       \n'
        result += '     ' + poschar[5]+'\n'
        result += '     ' + poschar[5]+'\n'
        result += '     ' + '\n'

    if number == 6:
        result += ' '+4*poschar[0]+' \n'
        result += poschar[1]+'     ' + '\n'
        result += poschar[1]+'     ' + '\n'
        result += ' '+4*poschar[3]+' \n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += ' '+4*poschar[5]+' \n'    

    if number == 5:
        result += ' '+4*poschar[0]+' \n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += ' '+4*poschar[3]+' \n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += ' '+4*poschar[5]+' \n'         

    if number == 4:
        result += '     '+' \n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += ' '+4*poschar[3]+' \n'
        result += '     ' + poschar[5]+'\n'
        result += '     ' + poschar[5]+'\n'
        result += '     ' + '\n'       

    if number == 3:
        result += ' '+4*poschar[0]+' \n'
        result += '     ' + poschar[2]+'\n'
        result += '     ' + poschar[2]+'\n'
        result += ' '+4*poschar[3]+' \n'
        result += '     ' + poschar[5]+'\n'
        result += '     ' + poschar[5]+'\n'
        result += ' '+4*poschar[5]+' \n' 

    if number == 2:
        result += ' '+4*poschar[0]+' \n'
        result += '     ' + poschar[2]+'\n'
        result += '     ' + poschar[2]+'\n'
        result += ' '+4*poschar[3]+' \n'
        result += poschar[4]+'     ' + '\n'
        result += poschar[4]+'     ' + '\n'
        result += ' '+4*poschar[5]+' \n'        

    if number == 1:
        result += '     '+' \n'
        result += '     ' + poschar[2]+'\n'
        result += '     ' + poschar[2]+'\n'
        result += '     '+' \n'
        result += '     ' + poschar[5]+'\n'
        result += '     ' + poschar[5]+'\n'
        result += '     '+' \n'

    if number == 0:
        result += ' '+4*poschar[0]+' \n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += poschar[1]+'    ' + poschar[2]+'\n'
        result += '     '+' \n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += poschar[4]+'    ' + poschar[5]+'\n'
        result += ' '+4*poschar[5]+' \n'        
    return result
```
