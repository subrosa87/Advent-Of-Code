```python
import numpy as np
import re
from collections import Counter
```

```python
file1 = open('input.txt', 'r')
#file1 = open('input.txt', 'r')
Lines = file1.readlines()
bingo_numbers_str = Lines[0].split(',')
bingo_numbers = [int(x) for x in bingo_numbers_str]
boards = []
board = []
board_line_count = 0
for i in range(2,len(Lines)):
    board_line_str = re.findall('\d+',Lines[i].replace('\n','')) 
    board_line_int = []
    for ele in board_line_str:
        board_line_int.append(int(ele))
    if(len(board_line_int)!=0): # and not i == len(Lines)-1):
        #print('if:',i,board_line_int)
        board.append(board_line_int)
    else:
        #print('else:',i,board_line_int)
        boards.append(np.array(board))
        board = []
    #if(i == len(Lines)-1):
    #    print('last if:',i,board_line_int)
    #    boards.append(np.array(board))
    #    board = []

boards.append(np.array(board))
print('boards:',boards)
```

```python
def evalBoard(bingo_board,bingo_numbers):
    bingo = False
    #print('board',bingo_board,'numbers',bingo_numbers)
    for row in bingo_board:
        bingo_count = 0
        cnt = Counter(row.tolist() + bingo_numbers)
        #print(cnt)
        for ele in cnt.values():
            if(ele > 1):
                bingo_count += 1
                if(bingo_count == 5):
                    bingo = True 
                    print('**** BINGO ****')
                    return [bingo_board,bingo_numbers] 
        #print(bingo_count,row)    
        bingo_count = 0            

    #print('--eval cols--')

    if not bingo:
        for i in range(0,5):
            bingo_count = 0
            col = bingo_board[:,i]
            cnt = Counter(col.tolist() + bingo_numbers)
            for ele in cnt.values():
                if(ele > 1):
                    bingo_count += 1 
                    if(bingo_count == 5):
                        bingo = True 
                        print('**** BINGO ****')
                        return [bingo_board,bingo_numbers] 
            #print(bingo_count,col)    
            bingo_count = 0    

    return None      
```

```python
def evalLoop(boards):
    result_exp = []

    winlose = []

    for board in boards:
        winlose.append(False)

    for i in range(5,len(bingo_numbers)):
        for k in range(0,len(boards)):
            if(not winlose[k]):
                ele = boards[k]
                result = evalBoard(ele,bingo_numbers[0:i])
                if not result == None: 
                    print(ele,type(ele),bingo_numbers[0:i])
                    winlose[k] = True 
                    result_exp.append(result)
    return result_exp
```

```python

result_a = evalLoop(boards) 
result = result_a[len(result_a)-1]

#result = evalBoard(boards[2],bingo_numbers[0:12])

print('result_arr       : ',result)
print('last number      : ',result[1][len(result[1])-1])

unmarked = []
for i in range(0,5):
    for k in range(0,5):
        if(not result[0][i][k] in result[1]):
            unmarked.append(result[0][i][k])



print('sum of unmarked  : ',sum(unmarked))
print('final result     : ',sum(unmarked) * result[1][len(result[1])-1])
```
