```python
def takeStep(m):
    flashCount = 0
    done = False
    m = m+1
    while not done:
        flash = np.where(m>=10)
        inc = []
        for x in range(0,m.shape[0]):
            for y in range(0,m.shape[1]):
                #inner
                if 0 < x < m.shape[0]-1 and 0 < y < m.shape[1]-1:
                    n = np.array([m[x+1,y],m[x-1,y],m[x,y-1],m[x,y+1],m[x+1,y+1],m[x-1,y-1],m[x+1,y-1],m[x-1,y+1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))
                #corner
                if 0 == x  and 0 == y :
                    n = np.array([m[x+1,y],m[x,y+1],m[x+1,y+1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))
                if m.shape[0]-1 == x  and m.shape[1]-1 == y :
                    n = np.array([m[x-1,y],m[x,y-1],m[x-1,y-1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))
                if 0 == x  and m.shape[1]-1 == y :
                    n = np.array([m[x+1,y],m[x,y-1],m[x+1,y-1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))
                if m.shape[0]-1 == x  and 0 == y :
                    n = np.array([m[x-1,y-1],m[x,y+1],m[x-1,y+1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))                    

                # edges 
                if x == 0 and 0 < y < m.shape[1]-1:
                    n = np.array([m[x+1,y],m[x+1,y-1],m[x+1,y+1],m[x,y-1],m[x,y+1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))              
                
                if x == m.shape[1]-1 and 0 < y < m.shape[1]-1:
                    n = np.array([m[x-1,y],m[x-1,y-1],m[x-1,y+1],m[x,y-1],m[x,y+1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))    

                if 0 < x < m.shape[0]-1 and y == 0:
                    n = np.array([m[x+1,y],m[x-1,y],m[x,y+1],m[x-1,y+1],m[x+1,y+1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))    

                if 0 < x < m.shape[0]-1 and y == m.shape[1]-1:
                    n = np.array([m[x+1,y],m[x-1,y],m[x,y-1],m[x+1,y-1],m[x-1,y-1]])
                    if len(n[n>=10]) > 0:
                        inc.append(((x,y),len(n[n>=10])))    

        flashCount += len(flash[0])
        #print('len flash:',len(flash[0]))
        for i in range(0,len(flash[0])):
            m[flash[0][i],flash[1][i]] = -99

        for i in inc:
            m[i[0]] += i[1]

        if len(flash[1]) == 0:# or runCount == 6:
            done = True 

    m[m < 0] = 0
    return (m,flashCount)
```
